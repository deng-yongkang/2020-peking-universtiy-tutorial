```python
# 绘制单条波形
def plotTrace(tr, starttime = 0, endtime= 86400):
    plt.figure(figsize=(10, 4))
    #plt.figure(figsize=(7, 3))
    depmax = np.max(tr.data)
    depmin = np.min(tr.data)
    maxY = np.fabs(depmin)
    if depmax > maxY:
        maxY = depmax
    myTimes = tr.times()
    myData = tr.data / depmax * 0.9 
    plt.plot(myTimes, myData, color='gray', linewidth=1)
    plt.xlim(starttime, endtime)
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.show()

# 时间域归一化，归一化方法选项可以为 lbit, clipping, clipping_iter, ramn
def normalize(tr, clip_factor=6, clip_weight=10, norm_win=10, norm_method="lbit"):
    if norm_method == 'clipping':
        lim = clip_factor * np.std(tr.data)
        tr.data[tr.data > lim] = lim
        tr.data[tr.data < -lim] = -lim
    elif norm_method == "clipping_iter":
        lim = clip_factor * np.std(np.abs(tr.data))

        while tr.data[np.abs(tr.data) > lim] != []:
            tr.data[tr.data > lim] /= clip_weight
            tr.data[tr.data < -lim] /= clip_weight
    elif norm_method == "ramn":
        lwin = int(tr.stats.sampling_rate * norm_win)
        st = 0
        N = lwin
        while N < tr.stats.npts:
            win = tr.data[st:N]
            w = np.mean(np.abs(win)) / (2. * lwin + 1)
            tr.data[st + lwin //2] /= w
            st += 1
            N += 1
          
        # 波形最前端以及最后端振幅逐渐衰减到 0，类似于上面
        taper = get_window(tr.stats.npts)
        tr.data *= taper
        
    elif norm_method == "lbit":
        tr.data = np.sign(tr.data)
        tr.data = np.float32(tr.data)
    return tr

def get_window(N, alpha=0.2):
    window = np.ones(N)
    x = np.linspace(-1., 1., N)
    ind1 = (abs(x) > 1 - alpha) * (x < 0)
    ind2 = (abs(x) > 1 - alpha) * (x > 0)
    window[ind1] = 0.5 * (1- np.cos(np.pi *(x[ind1]+1)/alpha))
    window[ind2] = 0.5 * (1- np.cos(np.pi *(x[ind2]-1)/alpha))
    # print(window)
    return window

# 谱白化
def whiten(tr, freqmin, freqmax):
    nsamp = tr.stats.sampling_rate
    n = len(tr.data)
    if n == 1:
        return tr
    else:
        frange = float(freqmax) - float(freqmin)
        nsmo = int(np.fix(min(0.01, 0.5 * (frange))* float(n)/nsamp))
        f = np.arange(n) * nsamp /(n -1.)
        JJ = ((f > float(freqmin)) & (f < float(freqmax))).nonzero()[0]

        # 信号的傅里叶变换
        FFTs = np.fft.fft(tr.data)
        FFTsW = np.zeros(n) + 1j * np.zeros(n)

        # 
        smo1 = (np.cos(np.linspace(np.pi / 2, np.pi, nsmo+1))**2)
        FFTsW[JJ[0]:JJ[0]+nsmo+1] = smo1 * np.exp(1j * np.angle(FFTs[JJ[0]:JJ[0]+nsmo+1]))

        FFTsW[JJ[0]+nsmo+1:JJ[-1]-nsmo] = np.ones(len(JJ) - 2 * (nsmo+1))\
        * np.exp(1j * np.angle(FFTs[JJ[0]+nsmo+1:JJ[-1]-nsmo]))

        smo2 = (np.cos(np.linspace(0., np.pi/2., nsmo+1))**2.)
        espo = np.exp(1j * np.angle(FFTs[JJ[-1]-nsmo:JJ[-1]+1]))
        FFTsW[JJ[-1]-nsmo:JJ[-1]+1] = smo2 * espo

        whitedata = 2. * np.fft.ifft(FFTsW).real
        
        tr.data = np.require(whitedata, dtype="float32")

        return tr

# 互相关
def correlateNoise(st, stations, corrwin):

    print ('correlating stations', (stations[0], stations[1]))

    # initialize sliding timewindow (length = corrwin) for correlation
    # start 1 corrwin after the start to account for different stream lengths
    timewin = st.select(station=stations[1])[0].stats.starttime + corrwin

    # loop over timewindows 
    # stop 1 corrwin before the end to account for different stream lengths
    while timewin < st.select(station=stations[0])[-1].stats.endtime - 2*corrwin:
        sig1 = st.select(station=stations[0]).slice(timewin, timewin+corrwin)
        sig1.merge(method=0, fill_value=0)
        sig2 = st.select(station=stations[1]).slice(timewin, timewin+corrwin)
        sig2.merge(method=0, fill_value=0)
        xcorr = np.correlate(sig1[0].data, sig2[0].data, 'same')

        try: 
            # build array with all correlations
            corr = np.vstack((corr, xcorr))
        except: 
            # if corr doesn't exist yet
            corr = xcorr
            
        # shift timewindow by one correlation window length
        timewin += corrwin

        # stack the correlations; normalize
        stack = np.sum(corr, 0)
        stack = stack / float((np.abs(stack).max()))    
    print ("...done")

    return corr, stack
# 叠加
def plotStack(st, stack, maxlag, figurename=None):

    # define the time vector for the correlation (length of corr = corrwin + 1)
    limit = (len(stack) / 2.) * st[0].stats.delta
    timevec = np.arange(-limit, limit, st[0].stats.delta)

    plt.plot(timevec, stack, 'k', lw=0.5)
    stations = list(set([_i.stats.station for _i in st]))
    plt.title("Stacked correlation between %s and %s" % (stations[0], stations[1]))
    plt.xlim(-maxlag, maxlag)
    plt.xlabel('time [s]')

    if figurename is not None:
        fig.savefig(figurename, format="pdf")
    else:
        plt.show()

def plotStack2(st, stack1, stack2 ,maxlag, figurename=None):
    limit = (len(stack1) / 2.) * st[0].stats.delta
    timevec = np.arange(-limit, limit, st[0].stats.delta)

    plt.plot(timevec, stack1, 'r', lw=1, alpha=0.7, label="Ref Waveform")
    plt.plot(timevec, stack2, 'k', lw = 0.5)
    stations = list(set([_i.stats.station for _i in st]))
    plt.title("Stacked correlation between %s and %s" % (stations[0], stations[1]))
    plt.xlim(-maxlag, maxlag)
    plt.legend(loc=1)
    plt.xlabel('time [s]')

    if figurename is not None:
        fig.savefig(figurename, format="pdf")
    else:
        plt.show()


# plot 实际Event 与互相关
def plotXcorrEvent(st, stn, stack, maxlag, acausal=False, figurename=None):

    eventtime = UTCDateTime(1998,7,15,4,53,21,0)                 # event near MLAC

    # station locations
    latP, lonP = 35.41, -120.55                                  # station PHL
    latM, lonM = 37.63, -118.84                                  # station MLAC
    latE, lonE = 37.55, -118.809                                 # event 1998
    
    # calculate distance between stations
    dist = gps2DistAzimuth(latP, lonP, latM, lonM)[0]            # between PHL and MLAC
    distE = gps2DistAzimuth(latP, lonP, latE, lonE)[0]           # between event and PHL
                                                                 #
    # CROSSCORRELATION
    # reverse stack to plot acausal part (= negative times of correlation)
    if acausal:
        stack = stack[::-1]
    
    # find center of stack
    c = int(np.ceil(len(stack)/2.) + 1)
    
    #cut stack to maxlag
    stack = stack[c - maxlag * int(np.ceil(stn[0].stats.sampling_rate)) : c + maxlag * int(np.ceil(stn[0].stats.sampling_rate))]
    
    # find new center of stack
    c2 = int(np.ceil(len(stack)/2.) + 1)

    # define time vector for cross correlation
    limit = (len(stack) / 2.) * stn[0].stats.delta
    timevec = np.arange(-limit, limit, stn[0].stats.delta)
    # define timevector: dist / t
    timevecDist = dist / timevec
    
    # EVENT
    ste = st.copy()
    st_PHL_e = ste.select(station='PHL')
    
    # cut down event trace to 'maxlag' seconds
    dt = len(stack[c2:])/stn[0].stats.sampling_rate                  #xcorrlength
    st_PHL_e[0].trim(eventtime, eventtime + dt)
    
    # create time vector for event signal
    # extreme values:
    limit = st_PHL_e[0].stats.npts * st_PHL_e[0].stats.delta
    timevecSig = np.arange(0, limit, st_PHL_e[0].stats.delta)

    # PLOTTING
    fig = plt.figure(figsize=(12.0, 6.0))
    ax1 = fig.add_subplot(2,1,1)
    ax2 = fig.add_subplot(2,1,2)

    # plot noise correlation
    ax1.plot(timevecDist[c2:], stack[c2:], 'k')
    ax1.set_title('Noise correlation between MLAC and PHL')

    # plot event near MLAC measured at PHL
    ax2.plot(distE/timevecSig, st_PHL_e[0].data / np.max(np.abs(st_PHL_e[0].data)), 'r')
    ax2.set_title('Event near MLAC observed at PHL')

    ax2.set_xlim((0, 8000))
    ax1.set_xlim((0, 8000))
    ax1.invert_xaxis()
    ax2.invert_xaxis()

    ax2.set_xlabel("group velocity [m/s]")
 
    if figurename is not None:
        fig.savefig(figurename, format="pdf")
    else:
        plt.show()        
```

### 模拟计算模块
这部分的模块主要利用的为上一节实验课中，模拟计算模块。

```python
def green_function_2d(t, x, speed=3):
    """
    2D格林函数：
    t：时间（s）
    x：距离（km）
    speed：速度（km/s）
    
    输出格林函数gf
    """
    t0 = x / speed
    gf = np.heaviside(t - t0, 0)
    idx = t > t0
    gf[idx] /= np.sqrt(t[idx]**2 - t0**2)
    
    return gf

def gf2d_int(t, x, c=3):
    t0 = x / c
    gf = np.heaviside(t - t0, 0)
    idx = t > t0
    gf[idx] = np.log((t[idx] + np.sqrt(t[idx]**2 - t0**2)) / t0)
    
    return gf

def ricker(t, t0=0, fc=.1, **kwargs):
    """
    雷克子波
    """
    tc = t - t0
    tmp = (np.pi * fc * tc)**2
    return (1 - 2*tmp) * np.exp(-tmp)

def impulse(t, t0=0, **kwargs):
    s = np.zeros(t.size)
    s[np.argmin(np.abs(t - t0))] = 1
    return s

def normalizeNum(a, axis=1):
    """
    标准化
    """
    if a.ndim == 1:
        return a / np.abs(a).max()
    else:
        return a / np.abs(a).max(axis=axis).reshape(-1, 1)

def _gf(time, source_radius, source_azimuth, receiver_location,
        source_time_function, gf_func=green_function_2d, c=3):
    """
    格林函数
    """
    if np.asarray(c).size == 1:
        c = np.full(source_azimuth.shape, c)

    d =  np.sqrt(
        (source_radius*np.cos(source_azimuth) - receiver_location)**2
        + (source_radius*np.sin(source_azimuth) - 0)**2
    )
    gf = []
    for _stf, _d, _c in zip(source_time_function, d, c):
        _gf = gf_func(time, _d, _c)
        gf.append(_gf)

    return np.vstack(gf)

def _displacement(source_time_function, green_function):
    """
    合成台站接收数据函数的内部调用
    """
    u = []
    for stf, gf in zip(source_time_function, green_function):
        u.append(normalizeNum(convolve(stf, gf, mode='full')[:gf.size]))
        
    return np.vstack(u)

def displacement(time, source_amplitude, source_time,
                 gf_func=green_function_2d, source_func=ricker,
                 source_radius=200, source_azimuth=np.arange(0, 2*np.pi, np.pi/360),
                 receiver_location_1=-100, receiver_location_2=100, c=3, **kwargs):
    """
    合成台站接收数据
    """
    stf = source_func(time, source_time.reshape(-1, 1))
    
    gf1 = _gf(time, source_radius, source_azimuth, receiver_location_1,
                 source_time_function=stf, gf_func=gf_func, c=c)
    u1 = source_amplitude.reshape(-1, 1) * _displacement(stf, gf1)
    gf2 = _gf(time, source_radius, source_azimuth, receiver_location_2,
                 source_time_function=stf, gf_func=gf_func, c=c)
    u2 = source_amplitude.reshape(-1, 1) * _displacement(stf, gf2)
    
    return stf, gf1, u1, gf2, u2

def displacement_direct(time,
                        source_func=ricker,
                        receiver_location_1=-100, receiver_location_2=100, c=3, shift=20, delta=.1, **kwargs):
    """
    格林函数计算得到的数据
    """
    stf = source_func(time, shift)
    ishift = int(shift / delta)
    stf = correlate(stf, stf, 'full')[-time.size-ishift:]
    x = abs(receiver_location_1 - receiver_location_2)
    gf = gf2d_int(time, x, c)
    u12 = convolve(stf, gf, mode='full')[ishift : gf.size+ishift-1]
    u = np.concatenate([-u12[::-1], [0], -u12])
    u *= np.hanning(u.size)

    return u

def cross_correlation(u1, u2):
    """
    互相关
    """
    xc = []
    for _u1, _u2 in zip(u1, u2):
        xc.append(correlate(_u1, _u2, 'full'))

    return np.vstack(xc)

def time_xc(t):
    return np.linspace(-t[-1], t[-1], 2*t.size-1)

def running_mean(a, n):
    """
    滑动平均
    """
    return convolve(a, np.ones(n)/n, mode='same')

def plt_source(source_amplitude, source_time,
                 source_radius=200, source_azimuth=np.arange(0, 2*np.pi, np.pi/360),
                 receiver_location_1=-100, receiver_location_2=100, step=10, **kwargs):
    """
    绘制震源图
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, subplot_kw={'projection': 'polar'}, figsize=(15, 10))
    
    # Schematic of geometry
    for _theta in [0, np.pi]:
        ax1.scatter( _theta, abs(receiver_location_1), marker='^',s=100)
    for theta in source_azimuth[::step]:
        ax1.scatter(theta, source_radius, marker='*')
    ax1.set_title("Source Station Location")
    # Source strength
    ax2.plot(source_azimuth, source_amplitude, lw=3)
    ax2.set_title('Source amplitude')
    plt.show()
#     axes[1].plot(theta, source_time, lw=3)
#     for ax in axes:
#         ax.set_theta_zero_location('N')
#         ax.set_theta_direction(-1)

    return ax1

def plt_waveform(u1, u2, xc, direct, source_azimuth, time,
                 xlim=400, step=10, **kwargs):
    """
    绘制波形图
    """
    fig, axes = plt.subplots(2, 3, gridspec_kw={'height_ratios': [.7, .3]}, figsize=(20, 10))

    for ax, u, t in zip(axes[0], [u1, u2, xc], [time, time, time_xc(time)]):
        vmax = np.abs(u).max()
        ax.pcolormesh(t[::step], np.rad2deg(source_azimuth)[::step], u[::step, ::step],
                      cmap='seismic', vmin=-vmax, vmax=vmax)
    
    xc_stack = xc.sum(axis=0)
    dt_xc = np.diff(xc_stack)
    mid = int(dt_xc.size / 2)
    dt_xc = np.concatenate([dt_xc[:mid], [0], dt_xc[mid:]])
    axes[1, 2].plot(time_xc(time), normalize(direct), c='k', ls='--', label='True')
    axes[1, 2].plot(time_xc(time), normalize(xc_stack), c='r', alpha=.5, label='Correlation stack')
    axes[1, 2].legend()

    for ax in axes[:, 2]:
        ax.set_xlim(-xlim/2, xlim/2)
    for ax in axes[0, :2]:
        ax.set_xlim(0, xlim)
    for ax in axes[1, :2]:
        ax.axis('off')
    axes[0, 0].set_ylabel(r'$\theta (^\circ)$')
    for ax, t in zip(axes[0, :], [r'$u_1$', r'$u_2$', r'$u_1 \star u_2$']):
        ax.set_title(t)
        ax.set_xlabel('Time (s)')
    for ax in axes[0, 1:]:
        ax.set_yticklabels([])
        
    return
def plt_waveform1(xc, direct, source_azimuth, 
                  time, xlim=400, step=10, **kwargs):
    xc_stack = xc.sum(axis=0)
    plt.figure(figsize=(10, 4))
    plt.plot(time_xc(time), normalizeNum(direct), c='k', ls='--', label='True')
    plt.plot(time_xc(time), normalizeNum(xc_stack), c='r', alpha=.5, label='Correlation stack')
    plt.legend()
    plt.show()
```
